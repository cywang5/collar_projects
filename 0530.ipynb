```python
import nltk
from nltk.corpus import PlaintextCorpusReader
from google.colab import drive
drive.mount('/content/drive')
corpus_root = '/content/drive/MyDrive/Colab Notebooks/Files'
wordlists = PlaintextCorpusReader(corpus_root, '.*')

books = wordlists.fileids()
print(books)
```

```python
import nltk
testText = wordlists.words('president.txt')
single = nltk.text.Text(testText)

findList = ["a", "the", "president", "Lai", "people"]
single.dispersion_plot(findList)
```

```python
books = wordlists.fileids()

cfd = nltk.ConditionalFreqDist((fileid, word)
  for fileid in books
  for word in wordlists.words(fileid)
)

findList = ['the', 'speech', 'elected', 'won', 'Taiwan', 'will', '2024']
cfd.tabulate(conditions = books, samples = findList)
```

```python
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:50]) #取至前50位字元
    for fileid in wordlists.fileids()
    for w in wordlists.words(fileid)
    for target in ('president', 'vice president', 'tsai', 'hsiao', 'lai')
    if w.lower() == target)
cfd.plot()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

#讀取文本
f = open('/content/drive/MyDrive/Colab Notebooks/Files/president.txt','r').read()

# stopwords 停用字詞
stop = ["bi", "will", "new", "s"] + list(STOPWORDS)

wordcloud = WordCloud(
    background_color="white",
    width=1500,
    height=1000,
    margin=2,
    stopwords=stop
).generate(f)


plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
import nltk

from nltk.corpus import PlaintextCorpusReader

corpus_root = '/content/drive/My Drive/Colab Notebooks/Files' # 請改成自己的路徑
wordlists = PlaintextCorpusReader(corpus_root, '.*')

books = wordlists.fileids()
print(books)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt


stop = list(STOPWORDS)

for fileid in books:
  f = open('/content/drive/MyDrive/Colab Notebooks/Files/'+fileid,'r').read()
  wordcloud = WordCloud(
      background_color="white",
      width=1500,
      height=1200,
      margin=2,
      stopwords=stop
  ).generate(f)

  plt.imshow(wordcloud)
  plt.axis("off")
  print(fileid)
  plt.show()
```
